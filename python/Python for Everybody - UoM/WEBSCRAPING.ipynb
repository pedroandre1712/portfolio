```python
from bs4 import BeautifulSoup
import urllib.request, urllib.parse, urllib.error
import re
```

Primeiro entregavel

```python
url = 'http://py4e-data.dr-chuck.net/comments_1478544.html'
html = urlopen(url).read()
soup = BeautifulSoup(html, 'html.parser')
print(soup.prettify())
```

```python
spans = soup('span')
sum = 0
for span in spans:
  match = re.findall('[0-9]+', str(span))
  sum+=int(match[0])
print(sum)
```

```python
url2 = 'http://py4e-data.dr-chuck.net/known_by_Clove.html'
html = urllib.request.urlopen(url2).read()
soup = BeautifulSoup(html, 'html.parser')
print(soup.prettify())
```

```python
def nome(url ,posicao, repeticao):
  if repeticao == 0:
    return ''
  html = urllib.request.urlopen(url).read()
  soup = BeautifulSoup(html, 'html.parser')
  tags = soup('a')
  anchor = tags[posicao-1]
  name = re.findall('([A-Z].+)\.',str(anchor))[0]
  new_url = re.findall('http.*html',str(anchor))[0]
  return (name + ' ' + nome(new_url, posicao, repeticao-1))
```

```python
nome(url2, 18, 7)
```

```python
for x in range(4):
  nome()
```
